### How to filter rows of df by column value


```python
movies.head()
```

```python
movies.shape
```

```python
booleans=[]
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
is_long=pd.Series(booleans)
```

```python
is_long.head
```

```python
movies[is_long]
```

```python
#method 2
is_long=movies.duration>=200
is_long.head()
movies[is_long]
```

```python
movies[movies.duration>=200]
```

```python
movies[movies.duration>=200].genre
```

```python
movies[movies.duration>=200]['genre']
```

```python
movies.loc[movies.duration>=200,'genre']
```

### Applying filter Criteria to a df

```python
movies[(movies.duration>=200) & (movies.genre=='Drama')]  #with multiple filtering criteria, separate by & (AND)
```

```python
movies[(movies.duration>=200) | (movies.genre=='Drama')] #with multiple filtering criteria, separate by | (OR)
```

```python
movies[(movies.duration>=200) | (movies.genre=='Drama')|(movies.genre=='Action')] #quite wordy
```

```python
movies[movies.genre.isin(['Crime','Drama','Action'])]
```

### Using Groupby in Pandas

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.max()
```

```python
drinks[drinks.continent=='Africa'].beer_servings.min()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
